```python
### What do you think about the students' parental level of education?
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(25, 10))
fig.suptitle('Comparing Parents education vs scores for all three subjects:', fontsize=20)

## math
sns.barplot(x='parental level of education', y='math score', data=data, ax=ax1)
ax1.title.set_text('Math score vs Parents education')
var1 = ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)

##reading
sns.barplot(x='parental level of education', y='reading score', data=data, ax=ax2)
ax2.title.set_text('Reading score vs Parents education')
var2 = ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)

## writing
sns.barplot(x='parental level of education', y='writing score', data=data, ax=ax3)
ax3.title.set_text('Writing score vs Parents education')
var3 = ax3.set_xticklabels(ax3.get_xticklabels(), rotation=45)
```

```python
### OBSERVATIONS
1, Students whose parents have Masters Degree scored the most in three subjects
2, Students whose parents have some high school degree scored the least in three subjects.
3, The parental level of education shows a clear impact on the student's score. The higher the parental education level, the better is the score of that student.
4, The order of students performance with respect to their parental education goes like this
   Masters > Bachelors > Associate > Some College > Some High School > High School.
```

```python
### What do you think about the scoring variation for math, reading and writing based on Test preparation course
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Comparing Test Preparation Course vs Scores for all three subjects:', fontsize=30)

## Math
sns.boxplot(x='test preparation course', y='math score', data=data, ax=ax1)
ax1.title.set_text('Math score vs Test Preparation Course')

## Reading
sns.boxplot(x='test preparation course', y='reading score', data=data, ax=ax2)
ax2.title.set_text('Reading score vs Test Preparation Course')

## Writing
sns.boxplot(x='test preparation course', y='writing score', data=data, ax=ax3)
ax3.title.set_text('Writing score vs Test Preparation Course')
```

```python
### OBSERVATIONS
#1, It is clear from these graphs that all the students who completed the course have scored better marks on an average than those who did not complete the course.
```

```python
### What do you think about the scoring variation for math, reading and writing based on gender.
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Comparing Male & Female scores for all three subjects:', fontsize=20) 

# Math
sns.boxplot(x='gender', y='math score', data=data, ax=ax1)
ax1.title.set_text('Math score vs Gender')

# Reading
sns.boxplot(x='gender', y='reading score', data=data, ax=ax2)
ax2.title.set_text('Reading score vs Gender')

# Writing
sns.boxplot(x='gender', y='writing score', data=data, ax=ax3)
ax3.title.set_text('Writing score vs Gender')
```

```python
### OBSERVATION
1, Male students scored more than females in maths and a female student scored zero in maths.
 2, The reading and writing score for female students is higher than male students, but the least score is obtained by a female student.
```
