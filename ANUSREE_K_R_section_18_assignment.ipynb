```python
Q1 = np.percentile(data['carat'],25, interpolation ='midpoint')
Q2 = np.percentile(data['carat'],50, interpolation ='midpoint')
Q3 = np.percentile(data['carat'],75, interpolation ='midpoint')
```

```python
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR = Q3 - Q1
```

```python
low_lim = Q1-1.5*IQR
up_lim = Q3+1.5*IQR
```

```python
print(low_lim)
print(up_lim)
```

```python
outlier = []
for x in data['carat']:
    if (x<low_lim) or (x>up_lim):
        outlier.append(x)
```
