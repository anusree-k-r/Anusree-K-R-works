Feature reduction

```python
#Removing features which are not relevant
data1 = data.drop(['post_age_category'],axis=1)
```

```python
## Remove the less important features
data2 = data1.drop(['Type','Date Posted','likes_per_day','comments_per_like'],axis=1)
```

```python
data2.columns
```

```python
data2.dtypes
```

```python
#Lets see the correlation between all variables in dataset
plt.figure(figsize=(14,10))
sns.heatmap(data2.corr(),annot=True,cmap='Greens',vmax=0.46,vmin=-0.98)
plt.show()
```

```python
#correlations of target column with other columns
plt.figure(figsize=(10,6))
corr=data1.corrwith(data1['Likes']).iloc[:-1].to_frame()
sns.heatmap(corr,cmap='Blues',annot=True)
plt.title('Heatmap of Target vs All columns',fontweight="bold", size=15)
plt.show()
```

```python
#Label encoding
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
label_enc=LabelEncoder()
```

```python
#label encoding Month, dayOfWeek, timeOfDay, Month
data2['Month']=label_enc.fit_transform(data2['Month'])
data2['dayOfWeek']=label_enc.fit_transform(data2['dayOfWeek'])
data2['timeOfDay']=label_enc.fit_transform(data2['timeOfDay'])
data2['Season']=label_enc.fit_transform(data2['Season'])
```

```python
#Standard Scaling of Likes Score
data_cont1=data2.select_dtypes(include = ['float'])
data_cont1.head()
```

```python
from sklearn import  preprocessing
std_scl=preprocessing.StandardScaler()
data_cont1=std_scl.fit_transform(data_cont1)
```

```python
data_cont1=pd.DataFrame(data_cont1,columns=['Likes Score'])
```

```python
data_cont1.shape
```

```python
data2.shape
```

```python
data2=data2.drop('Likes Score',axis=1)
```

```python
data2.shape
```

```python
data2=pd.concat([data2,data_cont1],axis=1)
```

```python
#Lets see the correlation between all variables in dataset after preprocessing
plt.figure(figsize=(20,15))
sns.heatmap(data2.corr(),annot=True,cmap='Greens',vmax=0.49,vmin=-0.98)
plt.show()
```
